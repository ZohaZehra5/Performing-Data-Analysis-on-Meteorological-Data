## Data Cleaning and Preparing


### Reading weatherHistory dataset through pandas and using info() function to get all the information.

```python
import pandas as pd
df = pd.read_csv("weatherHistory.csv")
df.info()
```

### Checking the missing entries

```python
df.isnull().sum()
```

This shows that only 'Precip type' has 517 null entries.

### Displaying the entries which have null value.


```python
df[df.isnull().any(axis=1)]
```

### Dropping the unnesscary columns as my required result is not dependent on these columns.

```python
df.drop(['Daily Summary' , 'Loud Cover'] , axis = 1 , inplace=True)
df
```

The column 'Daily Summary' was useless as we already have another column named as 'Summary' which gives us more precise detail per hourly where as in 'Daily Summary' it was constantly showing us the same result. 

I have also dropped 'Loud Cover' as it contained 0 value in all entries of the dataset.

### I'll be filling the missing entries in the original dataset.

```python
df.fillna('unknown' , inplace=True)
```

### Now confirming if the missing entries are filled now.

```python
df[df.isnull().any(axis=1)]
```

It does not show anything now so missing values are filled.

### Resampling the data from hourly to monthly

```python
df[['Formatted Date']]
```

```python
df['Formatted Date']=pd.to_datetime(df['Formatted Date'],utc=True)
df2=df.set_index('Formatted Date')
df2 = df2.resample('M').mean()
df2
```

UTC is Coordinated Universal Time. Followed univerisally.

## Now, performing data visualization using Seaborn & Matplotlib

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette("RdBu_r")
sns.set_style("whitegrid")

g = sns.relplot(x='Apparent Temperature (C)' , y='Humidity' , data=df2 , kind='line')
plt.figure(figsize=(12,8))
g.fig.suptitle("Apparent Temperature vs Humidity" , y=1.1)
plt.show()

from IPython.core.display import HTML
HTML("""
<style>
.output_png {
    display: table-cell;
    text-align: center;
    vertical-align: middle;
}
</style>
""")
```
